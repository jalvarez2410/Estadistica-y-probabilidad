## 1. Un semáforo localizado en cierta intersección está en verde 50% de las veces, en ámbar 10% y en rojo 40%. Un automóvil pasa por esta intersección una vez al día. Sea X el número de días que ha transcurrido, incluyendo la primera vez que el automóvil se topa con una luz roja. Suponga que cada día representa un experimento independiente.

* Determine P(X = 3).
* Determine P(X ≤ 3).
* Determine µX
* Determine σ




```python
# Probabilidad de encontrar la luz roja
p_rojo = 0.4

# 1. P(X = 3): Probabilidad de encontrar la luz roja por primera vez en el día 3
P_X_3 = geom.pmf(3, p_rojo)
print(f"P(X = 3): {P_X_3:.3f}")

# 2. P(X ≤ 3): Probabilidad acumulada hasta el día 3
P_X_leq_3 = geom.cdf(3, p_rojo)
print(f"P(X ≤ 3): {P_X_leq_3:.3f}")

# 3. Media (μ) de la distribución geométrica
mu_X = geom.mean(p_rojo)
print(f"Media (μₓ): {mu_X:.2f}")

# 4. Varianza (σ²) de la distribución geométrica
var_X = geom.var(p_rojo)
print(f"Varianza (σ²ₓ): {var_X:.2f}")
```

## 3 Un proceso que llena paquetes se detiene cada vez que se detecta uno cuyo peso no cumple la especificación. Suponga que cada paquete tiene probabilidad de 0.01 de no cumplir con la especificación y que los pesos de los paquetes son independientes.
* Determine el número promedio de paquetes que será llenado antes de que se detenga el proceso.
* Determine la varianza del número de paquetes que será Ilenado antes de que se detenga el proceso.
* Suponga que el proceso no se detendrá hasta que se detecten cuatro paquetes, cuyo peso no cumpla con la especificación. Determine la media y la varianza del número de paquetes que será llenado antes de que se detenga el proceso.

```python


# Probabilidad de que un paquete no cumpla con la especificación
p_defectuoso = 0.01

# a) Número promedio de paquetes llenados antes de detenerse (se resta 1 porque se considera antes del defecto)
media_geom = geom.mean(p_defectuoso) - 1
print(f"Número promedio de paquetes antes de detenerse: {media_geom:.2f}")

# b) Varianza del número de paquetes antes de detenerse
varianza_geom = geom.var(p_defectuoso)
print(f"Varianza del número de paquetes antes de detenerse: {varianza_geom:.2f}")

# c) Proceso que se detiene tras detectar 4 paquetes defectuosos (distribución binomial negativa)
r = 4
media_nbinom = nbinom.mean(r, p_defectuoso)
varianza_nbinom = nbinom.var(r, p_defectuoso)

print(f"Media del número de paquetes antes de detectar 4 defectuosos: {media_nbinom:.2f}")
print(f"Varianza del número de paquetes antes de detectar 4 defectuosos: {varianza_nbinom:.2f}")
```

## 4 El volumen de latas llenadas por cierta máquina se distribuye normalmente con media de 12.05 onzas y desviación estándar de 0.03 onzas.
* ¿Qué proporción de latas contiene menos de 12 onzas?
* La media del proceso se puede ajustar utilizando calibración. ¿En qué valor debe fijarse la media para que 99% de las latas contenga 12 onzas o más?
* Si la media del proceso sigue siendo de 12.05 onzas, ¿en qué valor debe fijarle la media para que 99% de las latas contenga 12 onzas o más?


```python
# Parámetros de la distribución
media = 12.05
desviacion = 0.03

# a) Proporción de latas con menos de 12 onzas
proporcion_menor_12 = norm.cdf(12, loc=media, scale=desviacion)
print(f"Proporción de latas con menos de 12 onzas: {proporcion_menor_12:.4f}")

# b) Ajuste de la media para que el 99% de las latas contenga 12 onzas o más
z_01 = norm.ppf(0.01)
nueva_media = 12 - z_01 * desviacion
print(f"Nueva media necesaria para que el 99% de las latas contengan al menos 12 onzas: {nueva_media:.4f}")

# c) Proporción de latas con 12 onzas o más si la media sigue siendo de 12.05 onzas
proporcion_mayor_12 = 1 - norm.cdf(12, loc=media, scale=desviacion)
print(f"Proporción de latas con 12 onzas o más con media de 12.05: {proporcion_mayor_12:.4f}")
```

## 5  Los ejes fabricados para el uso de dispositivos de almacenamiento ´optico tienen di´ametros que se distribuyen normalmente con media µ = 0.652 y desviaci´on est´andar σ = 0.003 cm . La especificaci´on para el di´ametro del eje mide entre 0.650 ± 0.005 cm.
* ¿Qu´e proporci´on de los ejes fabricados por este proceso cumple con la especificaci´on?
* La media del proceso puede ajustarse utilizando calibraci´on. Si se establece que la media mide 0.650 cm , ¿qu´e proporci´on de los ejes cumplir´a con la especificaci´on?
* Si se establece que la media mide 0.650 cm , ¿cu´al debe ser la desviaci´on est´andar para que 99% de los ejes cumpla con la especificaci´on?
Ayuda: Considere la siguiente propiedad de las distribuciones normales.
Sean X y Y independientes, con X ∼ N (µx, σ2x) y Y ∼ N(µY , σ2y), entonces
X + Y ∼ N (µX + µY , σ2x + σ2Y)
X − Y ∼ N (µX − µY , σ2X + σ2Y)


```python

# Parámetros de la distribución actual
media = 0.652
desviacion = 0.003

# a) Proporción de ejes que cumplen con la especificación (entre 0.645 y 0.655 cm)
limite_inferior = 0.645
limite_superior = 0.655

# Se calcula la probabilidad acumulada hasta el límite superior y se resta la acumulada hasta el límite inferior
proporcion_cumple = norm.cdf(limite_superior, loc=media, scale=desviacion) - norm.cdf(limite_inferior, loc=media, scale=desviacion)
print(f"Proporción de ejes que cumplen con la especificación (0.645 a 0.655 cm): {proporcion_cumple:.4f}")

# b) Proporción de ejes si la media se ajusta a 0.650 cm
nueva_media = 0.650
proporcion_cumple_ajustada = norm.cdf(limite_superior, loc=nueva_media, scale=desviacion) - norm.cdf(limite_inferior, loc=nueva_media, scale=desviacion)
print(f"Proporción de ejes con media ajustada a 0.650 cm: {proporcion_cumple_ajustada:.4f}")

# c) Cálculo de la desviación estándar necesaria para que el 99% de los ejes cumplan con la especificación
z = norm.ppf(0.995)


desviacion_necesaria = (limite_superior - nueva_media) / z
print(f"Desviación estándar necesaria para que el 99% de los ejes cumplan con la especificación: {desviacion_necesaria:.5f} cm")
```
